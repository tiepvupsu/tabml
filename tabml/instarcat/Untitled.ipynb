# Create a df with columns user_id, product_id
to represent that user_id ordered product_id sometime in the past

```python
df_orders = pd.read_csv("data/orders.csv")
df_orders.head(30)
```

```python
l1 = sorted(df["order_id"].unique().tolist())
```

```python
l2 = sorted(df_orders["order_id"].unique())
```

```python
in_both = []
only_in_l1 = []
j = 0
for order_id in l1:
    while j < len(l2) and order_id > l2[j]:
        j += 1
    if order_id == l2[j]:
        in_both.append(order_id)
    else:  # order_id < l2[j]
        only_in_l1.append(order_id)
```
